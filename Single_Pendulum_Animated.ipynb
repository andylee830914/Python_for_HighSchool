模擬單擺運動的重點=>求得角度與時間的關係=>解以下的二階常微分方程<br>
### $\left\{\begin{array}[rl]\displaystyle{\frac{d\theta}{dt}}&=\omega \\ \frac{d^2\theta}{dt^2}&=\frac{-g}{L}\sin\theta\end{array}\right.$

如果以下程式跑不出來，先安裝 [FFmpeg](https://www.ffmpeg.org)

```python
#REFERENCE http://matplotlib.org/examples/animation/double_pendulum_animated.html
from numpy import sin, cos
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate as integrate
import matplotlib.animation as animation

G = 9.80665  # acceleration due to gravity, in m/s^2
L = 1.0  # length of pendulum 1 in m
M = 1.0  # mass of pendulum 1 in kg

#dydx is a second order ode , $\frac{d^2\theta}{dt^2}=\frac{-g}{L}\sin\theta$
def derivs(state, t):

    dydx = np.zeros_like(state)
    dydx[0] = state[1] #angular velocities 
    dydx[1] = -G/L*sin(state[0])

    return dydx

# create a time array from 0..100 sampled at 0.05 second steps
dt = 0.05
t = np.arange(0.0, 20, dt)

# theta are the initial angles (degrees)
# omega are the initial angular velocities (degrees per second)
theta = 60.0
omega = 0.0

# initial state
state = np.radians([theta, omega])

# integrate your ODE using scipy.integrate.
y = integrate.odeint(derivs, state, t)

x1 = L*sin(y[:, 0])
y1 = -L*cos(y[:, 0])


fig = plt.figure()
ax = fig.add_subplot(111, autoscale_on=False, xlim=(-2, 2), ylim=(-2, 2))
ax.grid()
ax.set_aspect('equal', 'datalim')

line, = ax.plot([], [], 'o-', lw=2)
time_template = 'time = %.1fs'
time_text = ax.text(0.05, 0.9, '', transform=ax.transAxes)


def init():
    line.set_data([], [])
    time_text.set_text('')
    return line, time_text


def animate(i):
    thisx = [0, x1[i]]
    thisy = [0, y1[i]]
    line.set_data(thisx,thisy)
    time_text.set_text(time_template % (i*dt))
    return line, time_text

ani = animation.FuncAnimation(fig, animate, np.arange(1, len(y)),
                              interval=25, blit=True, init_func=init)

ani.save('single_pendulum.mp4', fps=20)
```
